```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Expanded_data_with_more_features.csv')
print(df.head())
```

```python
#Drop unnamed column

df= df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
df.head()
```

```python
#change weekly study hours column

df["WklyStudyHours"] = df["WklyStudyHours"].replace("05-10", "5-10")
df.head()
```

```python
#gender distribution

sns.countplot(data = df, x= "Gender")
plt.show()
```

```python
#Change figure size

plt.figure(figsize = (4,3))
sns.countplot(data = df, x= "Gender")
plt.show()
```

```python
#To know the exact count & label them

plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x= "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#From the above chart we have analysed that the number of females in the data is more than the number  
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
print(gb)
```

```python
#Heat Map

sns.heatmap(gb)
plt.show()
```

```python
#In order to show the exact value

plt.figure(figsize= (6,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

```python
#From the above chart we have concluded that:
The education of the parents have a good impact on their score
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
print(gb1)
```

```python
plt.figure(figsize = (6,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

```python
#From the above chart we have concluded that:
#There is no/negligible impact on the student's score due to their parent's marital status
```

```python
#Distribution of Ethnic Groups

groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["groupA", "groupB", "groupC","groupD","groupE"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]

plt.pie(mlist,labels = l)
plt.show()
```

```python
#Percentage Values 

groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["groupA", "groupB", "groupC","groupD","groupE"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
print(mlist)

plt.pie(mlist,labels = l, autopct = '%1.2f%%')  #Upto 2 decimal values of percentage
plt.title("Distribution of Ethnic Groups")
plt.show()
```
